```python
import json
import random
```

```python
train_path = "/content/sample_data/data.json"
with open(train_path, "r") as f:
    data = json.load(f)

print('Dataset size:', len(data))

random.shuffle(data)
print('Dataset is shuffled...')

train_dataset = data[:int(len(data)*0.9)]
test_dataset = data[int(len(data)*0.9):]

print(train_dataset)

print('Dataset is splitted...')
```

```python
import os
os.environ["TOKENIZERS_PARALLELISM"] = "true"

import torch
from gliner import GLiNERConfig, GLiNER
from gliner.training import Trainer, TrainingArguments
from gliner.data_processing.collator import DataCollatorWithPadding, DataCollator
from gliner.utils import load_config_as_namespace
from gliner.data_processing import WordsSplitter, GLiNERDataset
```

```python
device = torch.device('cuda:0') if torch.cuda.is_available() else torch.device('cpu')

model = GLiNER.from_pretrained("urchade/gliner_small")
```

```python
# use it for better performance, it mimics original implementation but it's less memory efficient
data_collator = DataCollator(model.config, data_processor=model.data_processor, prepare_labels=True)
```

```python
model.to(device)
print("done")
```

```python
# calculate number of epochs
num_steps = 500
batch_size = 16
data_size = len(train_dataset)
num_batches = data_size // batch_size
num_epochs = max(1, num_steps // num_batches)

training_args = TrainingArguments(
    output_dir="models",
    learning_rate=1e-5,
    weight_decay=0.01,
    others_lr=1e-5,
    others_weight_decay=0.01,
    lr_scheduler_type="linear", #cosine
    warmup_ratio=0.1,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    focal_loss_alpha=0.75,
    focal_loss_gamma=2,
    num_train_epochs=num_epochs,
    evaluation_strategy="steps",
    save_steps = 100,
    save_total_limit=10,
    dataloader_num_workers = 4,
    use_cpu = False,
    report_to="none",
    )

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=model.data_processor.transformer_tokenizer,
    data_collator=data_collator,
)

trainer.train()
```

```python
trained_model = GLiNER.from_pretrained("/content/models/checkpoint-750", load_tokenizer=True)
trained_model.save_pretrained("/content/models/checkpoint-750")
```

```python
labels = ["Configuration", "Area", "Developer", "Price", "Location"]

test_examples = [
"Modern 3BHK apartment with garden view, 1650 sq.ft built-up area by Raheja Developers, priced at ₹1.85Cr in Powai, Mumbai.",
"Elegant 2BHK with 1100 sq.ft carpet area by Hiranandani Group for ₹1.4Cr in Thane West.",
"Premium 4BHK penthouse with 2800 sq.ft super built-up area by Sobha Limited, available at ₹3.2Cr in Whitefield, Bangalore.",
"Affordable 1BHK of 620 sq.ft by Shapoorji Pallonji priced at ₹65 lacs in Ghodbunder Road, Thane.",
"Luxurious 5BHK duplex with 3200 sq.ft area by Emaar Properties for ₹4.8Cr in Sector 65, Gurgaon.",
"Sea-facing 3BHK of 1750 sq.ft by Kalpataru Group for ₹2.3Cr in Worli, Mumbai.",
"Budget 2BHK apartment with 950 sq.ft built-up area by Vatika Group, priced at ₹78 lacs in Greater Noida.",
]

# Process each example
for i, text in enumerate(test_examples):
    print(f"\nExample {i+1}: {text}")

    # Predict entities
    entities = trained_model.predict_entities(text, labels, threshold=0.05)

    # Display results
    print("\nPredicted Entities:")
    for entity in entities:
        print(f"{entity['text']} => {entity['label']}")

    print("-" * 50)
```
